Import necessary libraries

```python
import pandas as pd
import numpy as np
```

**1.** Fetch dataset. In order to have possibility to fetch it by URL, I have uploaded dataset to my GitHub.

Dataset is about aid from different countries for Ukraine. Original dataset is taken from https://www.kaggle.com/datasets/jainaru/financial-aid-to-ukraine.

```python
df = pd.read_csv("https://raw.githubusercontent.com/YaroslavDrovolskyi/applied-application-of-neural-networks/main/lab1/Financial%20Allocations.csv")
```

**2.** Print dataset

```python
df
```

**3.** Check if any value in dataframe is `NaN`.

```python
df.isnull().any().any()
```

**4.** Dataset has `NaN` values, so need to replace them by `10`. Print dataset with replaces `NaN`s

```python
df.fillna(10)
```

**5.** Delete rows with `NaN` values. (Drop the rows where at least one element is `NaN`).

```python
df=df.dropna()
df
```

**6.** Create new column ('Total allocations')  as a sum of three existing columns

```python
df.loc[:, 'Total allocations'] = df['Financial allocations($ billion)'] + df['Humanitarian allocations($ billion)'] + df['Military allocations($ billion)']
```

```python
df
```

**7.** Make generalized description of dataset

```python
df.describe()
```

As we see, each country, **on average**, has allocated 4.895068 bn USD for Ukraine as support.

Minimum donation is 0.002218 bn USD, maximal is 71.858826 bn USD.


**8.** Make analysis for EU and non-EU countries






```python
df.groupby(['EU member']).agg({'Financial allocations($ billion)':[min, max, sum, 'mean'],
                               'Humanitarian allocations($ billion)':[min, max, sum, 'mean'],
                               'Military allocations($ billion)':[min, max, sum, 'mean'],
                               'Total allocations':[min, max, sum, 'mean']})
```

As we see, average non-EU country has donated 8.2 bn USD for Ukraine, and average EU country - 3.38 bn USD.

Minimum not-EU country allocation is 0.002284 bn USD, maximal is 71.858826 bn USD.
Minimum EU country allocation is 0.002218 bn USD, maximal is 32.106644 bn USD.

So, EU countries have allocated 81.12 bn USD, but not-EU countries - 90.2 bn USD.
