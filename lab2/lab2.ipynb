Зобразимо 4 графіка:
- ***Графік популяції за 1951-2020 роки***. На ньому видно, що популяція Землі зростала з кожним роком
- ***Графік зміни популяції порівняно з попереднім роком (у %) за 1951-2020 роки***. На ньому видно, що:
  - ця зміна завжди була додатньою (а це означає, що наслення Земі завди зростало)
  - графік зміни - нелінійний, і в останні роки спостерігаємо спад темпу збільшення кількості населення (до менше ніж 0.5% (для порівняння, у 1970-х роках було більше ніж 2%))
- ***Графік густоти населення з 1951 по 2020 рік***. Бачимо, що густота постійно збільшується, причому зростання приблизно лінійне. За 70 років спостережень густота більшилась більше ніж у 3.7 рази.
- ***Графік популяції разом із графіком міського населення.*** Обидва графіки зротстають. За 7 років популяція зроосла у 3 рази, а популяція у містах - у 5 разів. Темпи збільшення зараз приблизно однакові, хоча ще у 1980-1990-х роках спостерігалось незначне уповільнення (увігнутість графіка вниз) темпу зростання міського населення


```python
plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.plot(df_population["Year"], df_population["Population"])
plt.title("Population on Earth")
plt.xlabel("Year")
plt.ylabel("Population")

plt.subplot(2,2,2)
plt.plot(df_population["Year"], df_population["ChangePerc"], "go")
plt.title("Change (%) of population relative to previous year")
plt.xlabel("Year")
plt.ylabel("Change, %")

plt.subplot(2,2,3)
plt.plot(df_population["Year"], df_population["Density"], "r^")
plt.title("Density of population")
plt.xlabel("Year")
plt.ylabel("Density")

plt.subplot(2,2,4)
plt.plot(df_population["Year"], df_population["Population"], ".",
         df_population["Year"], df_population["Urban"], "s")
plt.title("Population on Earth and Urban population")
plt.xlabel("Year")
plt.ylabel("Population and Urban population")
plt.legend(['Population', 'Urban population'])

# margins between subplots
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)


plt.show()
```

Будуємо стовпчату та кругову діаграми для категорій (фінансова, гуманітарна, військова) допомоги Україні сумарно з усіх країн.

Як видно з двох графіків, найменшою є частка гуманітарної допомоги (~15 млрд. дол. США), найбільша - військової допомоги (> 80 млрд. дол. США).

```python
plt.figure(figsize=(15,5))

allocation_types_names = ['Financial', 'Humanitarian', 'Military']
allocation_types_values = [df_ukr_support['Financial allocations($ billion)'].sum(),
                           df_ukr_support['Humanitarian allocations($ billion)'].sum(),
                           df_ukr_support['Military allocations($ billion)'].sum()]

# bar chart
plt.subplot(1,2,1)
plt.bar(allocation_types_names, allocation_types_values, color="lightblue")
plt.rc('axes', axisbelow=True, edgecolor='gray') # grid under bars, and plot frame to be gray
plt.grid()
plt.title("Types of allocations for Ukraine")
plt.xlabel("Allocation type")
plt.ylabel("Billions of USD")

# pie chart
plt.subplot(1,2,2)
plt.pie(allocation_types_values, labels=allocation_types_names,
        wedgeprops={'linewidth': 1.5, 'edgecolor':'white'}) # wedgeprops is lines between pie parts, https://stackoverflow.com/a/73040402
plt.axis('equal')
plt.legend(title='Allocation types')
plt.title("Types of allocations for Ukraine")

plt.show()
```

Додамо в датасеті стовпчик "Десятиліття", для коректної побудови наступних графіків.

```python
from math import trunc

year_by_10_frac, year_by_10_int = np.modf(df_population['Year'] / 10)

df_population.loc[:, 'Decade'] = year_by_10_int.astype(int) * 10
```

Будуємо діаграму розсіювання (scatter plot), лінійний графік (line plot) та стовпчату діаграму (bar plot) відносної зміни популяції на Землі порівняно з попереднім роком (стовпчик ChargePerc)

*(Сторінка 7 з презентації)*

```python
# How to make subplots with sns: https://stackoverflow.com/a/41384984
# Size of sns plots: https://stackoverflow.com/a/31597278

figure, axes = plt.subplots(1, 3, figsize=(18,5))
sns.scatterplot(data=df_population, x="Decade", y="ChangePerc", ax=axes[0]).set(title="Population change", ylabel="Population change, %")
sns.lineplot(data=df_population, x="Decade", y="ChangePerc", ax=axes[1]).set(title="Population change", ylabel="Population change, %")
sns.barplot(data=df_population, x="Decade", y="ChangePerc", ax=axes[2],
            palette=sns.color_palette("husl", 8), hue="Decade", legend=False).set(title="Population change", ylabel="Population change, %")
            # palette is for coloring histograms
```

Як видно із графіків, протягом 1951-2020 років, кількість населення завжди зростала (ChangePerc > 0). Пік росту був у 1960-х роках, далі поступово зменшувався, і станом на зараз досяг свого мінімуму - 1.05%.

На лінійновму графіку (центральний графік) також зображено середнє значення та 95% довірчий інтервал для значень в межах одного й того самого десятиліття.

Стовпчата діаграма (праворуч) зображає агреговану оцінку ChangePerc для кожного десятиліття за допомогою висоти прямокутників, та вказує на похибку оцінки.

Як густота населення залежала від десятиліття та збільшення кількості населення в період 1951-2020 років?

*Слайд 11 з презентації*

```python
pivot = df_population.pivot_table(
    index=["ChangePerc"],
    columns=["Decade"],
    values="Density",
    aggfunc=np.average)
sns.heatmap(pivot).set(title="Density", xlabel="Decade", ylabel="Change of population, %")
```

З графіку бачимо, що чим менше зростання населення у відсотках та чим ближче десятиліття до сучасності - тим густоста населення більша.
